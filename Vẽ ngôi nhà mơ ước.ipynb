```python
import turtle

pen = turtle.Turtle()
turtle.bgcolor("lightblue")
#tăng tốc
pen.speed(50)

# hình vuông
pen.fillcolor ("orange")
pen.begin_fill()
pen.left(130)
pen.forward(150)
pen.left(90)
pen.forward(150)
pen.left(90)
pen.forward(150)
pen.left(90)
pen.forward(150)
pen.end_fill()

# hình thang
pen.fillcolor ("yellow")
pen.begin_fill()
pen.right(130)
pen.forward(250)
pen.right(60)
pen.forward(133)
pen.right(120)
pen.forward(220)
pen.end_fill()

# hình tam giác
pen.fillcolor ("purple")
pen.begin_fill()
pen.left(90)
pen.forward(200)
pen.right(133)
pen.forward(160)
pen.end_fill()
pen.bk(160)

# hình chữ nhật
pen.fillcolor ("blue")
pen.begin_fill()
pen.lt(43)
pen.bk(220)
pen.rt(90)
pen.fd(200)
pen.lt(90)
pen.fd(220)
pen.lt(90)
pen.fd(200)
pen.end_fill()


pen.penup()
pen.goto(-260,-207)
pen.down()
# cửa
pen.fillcolor ("lightgreen")
pen.begin_fill()
pen.lt(90)
pen.fd(110)
pen.lt(90)
pen.fd(80)
pen.lt(90)
pen.fd(110)
pen.end_fill()

pen.penup()
pen.goto(-110,-150)
# cửa sổ
pen.fillcolor ("brown")
pen.begin_fill()
pen.right(60)
pen.forward(40)
pen.pendown()
pen.forward(50)
pen.right(120)
pen.forward(75)
pen.right(60)
pen.forward(50)
pen.right(120)
pen.forward(75)
pen.end_fill()

#Cây 1
pen.penup()
pen.goto(250,0)
pen.pendown()
pen.fillcolor ("lightgreen")
pen.begin_fill()
pen.left(135)
pen.forward(100)
pen.left(135)
pen.forward(150)
pen.left(140)
pen.forward(110)
pen.end_fill()

#Cây 2
pen.fillcolor ("lightgreen")
pen.begin_fill()
pen.penup()
pen.goto(250,-70)
pen.pendown()
pen.right(50)
pen.left(135)
pen.forward(100)
pen.left(135)
pen.forward(150)
pen.left(140)
pen.forward(110)
pen.end_fill()

#Cây 3
pen.fillcolor ("lightgreen")
pen.begin_fill()
pen.penup()
pen.goto(250,-140)
pen.pendown()
pen.right(50)
pen.left(135)
pen.forward(100)
pen.left(135)
pen.forward(150)
pen.left(140)
pen.forward(110)
pen.end_fill()

#Thân
pen.fillcolor ("brown")
pen.begin_fill()
pen.penup()
pen.goto(230,-210)
pen.pendown()
pen.left(130)
pen.forward(75)
pen.left(90)
pen.forward(50)
pen.left(90)
pen.forward(75)
pen.left(90)
pen.forward(50)
pen.end_fill()

#Mặt trời
pen.penup()
pen.goto(150,300)
pen.pendown()

pen.pensize (3)
 
pen.pencolor ("black")

pen.fillcolor ("yellow")
pen.begin_fill()
pen.circle (50, extent = 45)
pen.right(90)
pen.fd(25)
pen.backward(25)
pen.left(90)
pen.circle (50, extent = 45)
pen.right(90)
pen.fd(50)
pen.backward(50)
pen.left(90)
pen.circle (50, extent = 45)
pen.right(90)
pen.fd(25)
pen.backward(25)
pen.left(90)
pen.circle (50, extent = 45)
pen.right(90)
pen.fd(50)
pen.backward(50)
pen.left(90)
pen.circle (50, extent = 45)
pen.right(90)
pen.fd(25)
pen.backward(25)
pen.left(90)
pen.circle (50, extent = 45)
pen.right(90)
pen.fd(50)
pen.backward(50)
pen.left(90)
pen.circle (50, extent = 45)
pen.right(90)
pen.fd(25)
pen.backward(25)
pen.left(90)
pen.circle (50, extent = 45)
pen.right(90)
pen.fd(50)
pen.backward(50)
pen.left(90)
pen.end_fill()

turtle.done()
```
