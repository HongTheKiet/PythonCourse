```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
import openpyxl

#tạo file excel
wb = openpyxl.Workbook()
sheet = wb.active

# Đường dẫn tới file Excel
file_path = "orders.xlsx"

# Tạo workbook từ file Excel
workbook = openpyxl.load_workbook(file_path)

# Lấy danh sách các sheet trong workbook
sheet_names = workbook.sheetnames

# Chọn sheet đầu tiên
sheet = workbook[sheet_names[0]]

# Lấy tất cả các dòng trong sheet
rows = sheet.iter_rows(min_row=2, values_only=True)

# Hiển thị thông tin từ file Excel
for row in rows:
    date = row[0]
    order_id = row[1]
    customer_name = row[2]
    product_id = row[3]
    
    print(f"Ngày nhận đơn: {date}, Mã đơn hàng: {order_id}, Tên khách hàng: {customer_name}, Mã sản phẩm: {product_id}")

# Đóng workbook
workbook.close()

#tạo cửa sổ tkinter
window = Tk()
window.title("Quản lý đơn hàng")
window.geometry("650x400")

# Tạo các hộp văn bản để nhập thông tin đơn hàầy
#Label
date_label = Label(window, text="Ngày nhận đơn:")
order_id_label = Label(window, text="Mã đơn hàng:")
customer_name_label = Label(window, text="Tên khách hàng:")
product_id_label = Label(window, text="Mã sản phẩm:")

#Entry
date_entry = Entry(window)
order_id_entry = Entry(window)
customer_name_entry = Entry(window)
product_id_entry = Entry(window)

# Tạo danh sách đơn hàng sử dụng Treeview
order_treeview = ttk.Treeview(window)
order_treeview["columns"] = ("Date", "OrderID", "CustomerName", "ProductID")
order_treeview.column("#0", width=0, stretch=NO)
order_treeview.column("Date", width=100)
order_treeview.column("OrderID", width=100)
order_treeview.column("CustomerName", width=100)
order_treeview.column("ProductID", width=200)

order_treeview.heading("Date", text="Ngày nhận đơn")
order_treeview.heading("OrderID", text="Mã đơn hàng")
order_treeview.heading("CustomerName", text="Tên khách hàng")
order_treeview.heading("ProductID", text="Mã sản phẩm")

# Tạo scrollbar cho Treeview
scrollbar = ttk.Scrollbar(window, orient="vertical", command=order_treeview.yview)
order_treeview.configure(yscrollcommand=scrollbar.set)

# Tạo các nút điều khiển
create_button = Button(window, text="Tạo đơn hàng")
find_button = Button(window, text="Tìm đơn hàng")
delete_button = Button(window, text="Xóa đơn hàng")
update_button = Button(window, text="Cập nhật đơn hàng")
save_button = Button(window, text="Lưu dữ liệu", command=save_to_excel)

date_label.grid(row=0, column=0)
date_entry.grid(row=0, column=1)

order_id_label.grid(row=1, column=0)
order_id_entry.grid(row=1, column=1)

customer_name_label.grid(row=2, column=0)
customer_name_entry.grid(row=2, column=1)

product_id_label.grid(row=3, column=0)
product_id_entry.grid(row=3, column=1)

order_treeview.grid(row=4, columnspan=2, sticky=(N, S, E, W))
scrollbar.grid(row=4, column=2, sticky=(N, S))

create_button.grid(row=5, column=0)
find_button.grid(row=5, column=1)
delete_button.grid(row=6, column=0)
update_button.grid(row=6, column=1)
save_button.grid(row=7, column=0, columnspan=2)

def create_order():
    # Lấy thông tin từ các hộp văn bản nhập liệu
    date = date_entry.get()
    order_id = order_id_entry.get()
    customer_name = customer_name_entry.get()
    product_id = product_id_entry.get()

    # Thêm đơn hàng vào Treeview
    order_treeview.insert("", END, text="", values=(date, order_id, customer_name, product_id))

    # Xóa nội dung các hộp văn bản nhập liệu
    date_entry.delete(0, END)
    order_id_entry.delete(0, END)
    customer_name_entry.delete(0, END)
    product_id_entry.delete(0, END)

def find_order():
    # Lấy thông tin từ các hộp văn bản nhập liệu
    order_id = order_id_entry.get()

    # Tìm kiếm đơn hàng trong Treeview
    for child in order_treeview.get_children():
        order = order_treeview.item(child)["values"]
        if order[1] == order_id:  # Check the OrderID column value
            messagebox.showinfo("Tìm đơn hàng", f"Ngày nhận đơn: {order[0]}, Mã đơn hàng: {order[1]}, Tên khách hàng: {order[2]}, Mã sản phẩm: {order[3]}")
            break
    else:
        messagebox.showinfo("Tìm đơn hàng", "Không tìm thấy đơn hàng")

def delete_order():
    # Lấy thông tin từ các hộp văn bản nhập liệu
    order_id = order_id_entry.get()

    # Tìm kiếm và xóa đơn hàng trong Treeview
    for child in order_treeview.get_children():
        order = order_treeview.item(child)["values"]
        if order[1] == order_id:  # Check the OrderID column value
            order_treeview.delete(child)
            messagebox.showinfo("Xóa đơn hàng", "Đã xóa đơn hàng thành công")
            break
    else:
        messagebox.showinfo("Xóa đơn hàng", "Không tìm thấy đơn hàng")

def update_order():
    # Lấy thông tin từ các hộp văn bản nhập liệu
    order_id = order_id_entry.get()

    # Tìm kiếm và cập nhật đơn hàng trong Treeview
    for child in order_treeview.get_children():
        order = order_treeview.item(child)["values"]
        if order[1] == order_id:  # Check the OrderID column value
            updated_order = (date_entry.get(), order_id, customer_name_entry.get(), product_id_entry.get())
            order_treeview.item(child, values=updated_order)
            messagebox.showinfo("Cập nhật đơn hàng", "Đã cập nhật đơn hàng thành công")
            break
    else:
        messagebox.showinfo("Cập nhật đơn hàng", "Không tìm thấy đơn hàng")

def select_order(event):
    # Lấy thông tin của đơn hàng đã chọn
    selected_item = order_treeview.focus()
    values = order_treeview.item(selected_item, "values")

    # Hiển thị thông tin của đơn hàng lên các Entry
    date_entry.delete(0, END)
    date_entry.insert(END, values[0])

    order_id_entry.delete(0, END)
    order_id_entry.insert(END, values[1])

    customer_name_entry.delete(0, END)
    customer_name_entry.insert(END, values[2])

    product_id_entry.delete(0, END)
    product_id_entry.insert(END, values[3])

def save_to_excel():
    # Create a new workbook
    workbook = openpyxl.Workbook()
    sheet = workbook.active

    # Write the headers
    headers = ["Ngày nhận đơn", "Mã đơn hàng", "Tên khách hàng", "Mã sản phẩm"]
    for idx, header in enumerate(headers, start=1):
        sheet.cell(row=1, column=idx, value=header)

    # Write the data from the Treeview to the worksheet
    values = order_treeview.get_children()
    for row, item in enumerate(values, start=2):
        data = order_treeview.item(item)["values"]
        for col, value in enumerate(data, start=1):
            sheet.cell(row=row, column=col, value=value)

    # Save the workbook
    workbook.save("orders.xlsx")
    messagebox.showinfo("Lưu trữ", "Dữ liệu đã được lưu trữ vào file Excel.")
  
# Gắn sự kiện cho Treeview để chọn và cập nhật đơn hàng
order_treeview.bind("<<TreeviewSelect>>", select_order)

# Gắn sự kiện cho nút "Tạo đơn hàng"
create_button.configure(command=create_order)

# Gắn sự kiện cho nút "Tìm đơn hàng"
find_button.configure(command=find_order)

# Gắn sự kiện cho nút "Xóa đơn hàng"
delete_button.configure(command=delete_order)

# Gắn sự kiện cho nút "Cập nhật đơn hàng"
update_button.configure(command=update_order)

window.protocol("WM_DELETE_WINDOW", on_closing)
# Chạy vòng lặp chính của ứng dụng
window.mainloop()
```
